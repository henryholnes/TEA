```python
import pandas as pd
from matplotlib import pyplot as plt
datos=pd.read_csv("datos.csv")
datos
quimia=datos[datos.m=="Quimia"]
quimia
cosmocel=datos[datos.m=="Cosmocel"]
cosmocel
plt.style.use('dark_background')

plt.plot(quimia.n, quimia.precio)
plt.plot(cosmocel.n, cosmocel.precio)
plt.legend(["Rainbow", "Cosmocel"])
plt.xlabel("Numero")
plt.ylabel("Nombre")
plt.rcParams['figure.figsize']=[2,3]
plt.show()
```
